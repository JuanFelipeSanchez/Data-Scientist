A. Extraer la información del archivo.

```python
#Se abre el archivo y se imprime el Dataframe.
Vacunas = pd.read_csv('/content/gdrive/MyDrive/Data Scientist/5. Panda y Numpy en Python/country_vaccinations.csv')
print(Vacunas)
```

B. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
#Se imprime la estructura y los tipos de datos de cada columna del Dataframe verificando el formato de las fechas.
Vacunas['date']=pd.to_datetime(Vacunas['date'])
Vacunas.info()
```

C. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
#Se calcula el total de vacunas aplicadas "total_vaccinations" por cada compañía, agrupando por la columna "vaccines" y sumando los resultados.
resultado_agrupación = Vacunas[['vaccines', 'total_vaccinations']].groupby('vaccines').sum()

#El resultado es generado con la estructura solicitada, pero la sumatoria se refleja en formato de notación científica.
#Aunque este formato puede ser permitido, en este caso, se usa la función map para cambiar a formato Int el resultado de este cálculo.
resultado_agrupación = resultado_agrupación.map(lambda x: int(x))
print(resultado_agrupación)
```

D. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
#Se calcula el total de vacunas aplicadas en todo el mundo.
Total_Vacunas = Vacunas['total_vaccinations'].sum()
Total_Vacunas = int(Total_Vacunas)
print('El total de vacunas aplicadas en todo el mundo es: ',Total_Vacunas)
```

E. Calcular el promedio de vacunas aplicadas por país.

```python
#Se calcula el promedio de vacunas aplicadas por país, agrupando por la columna "country" y generando los cálculos.
Promedio_vacunas = Vacunas[['country', 'total_vaccinations']].groupby('country').mean()

#El resultado es generado con la estructura requerida, pero el promedio se refleja en formato de notación científica.
#Aunque este formato puede ser permitido, en este caso, se usa la función map para cambiar a formato decimal el resultado de este cálculo.
Promedio_vacunas = Promedio_vacunas.map(lambda x: f"{x:.1f}")
print(Promedio_vacunas)
```

F. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
#Se calcula la cantidad de vacunas que han sido aplicadas en el día 29/01/21 en todo el mundo.

#1. Se crea un nuevo Dataframe con los registros efectuados el día 29/01/21.
Vacunas_Fecha_específica = Vacunas[Vacunas['date']=='2021-01-29']

#2. Se realiza la sumatoria de vacunas aplicadas con los datos del nuevo Dataframe.
Total_Vacunas_Fecha_específica = Vacunas_Fecha_específica['total_vaccinations'].sum()
Total_Vacunas_Fecha_específica = int(Total_Vacunas_Fecha_específica)
print('El total de vacunas aplicadas en el día 29/01/21 en todo el mundo es: ',Total_Vacunas_Fecha_específica)
```

G. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
#Se crea un nuevo Dataframe llamado conDiferencias que tiene los datos originales y una nueva columna con las diferencias de aplicación entre las columnas
#daily_vaccionations y daily_vaccionations_raw.

#1. Se crea el Dataframe conDiferencias.
conDiferencias = pd.read_csv('/content/gdrive/MyDrive/Data Scientist/5. Panda y Numpy en Python/country_vaccinations.csv')

#2. Se crea la nueva columna con las diferencias. Se usa la función fillna(0) para que las diferencias puedan realizarse en caso de que hayan valores nulos.
#Si se quiere omitir esto solo se debe eliminar dicha función.
conDiferencias['daily_vaccinations_raw'] = conDiferencias['daily_vaccinations_raw'].fillna(0)
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
```

H. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
#Se calcula el periodo de tiempo entre el registro más reciente al más antiguo.

#1. Obtención del registro más reciente.
Reciente = max(Vacunas['date'])

#2. Obtención del registro más antiguo.
Antiguo = min(Vacunas['date'])

#3. Se restan para obtener el periodo de tiempo transcurrido entre una fecha y la otra.
Periodo = Reciente - Antiguo
Periodo = Periodo.days

print("El periodo de tiempo entre el registro más reciente al más antiguo equivale a ",Periodo," días")
```

I. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
#Se crea el Dataframe conCantidad, igual al original, con una columna nueva con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

#1. Se crea el nuevo Dataframe.
conCantidad = pd.read_csv('/content/gdrive/MyDrive/Data Scientist/5. Panda y Numpy en Python/country_vaccinations.csv')

#2. Creación de nueva columna con la cantidad de tipos de vacunas usadas cada día.
conCantidad['canVac'] = conCantidad['vaccines'].apply(lambda x: len(x.split(',')))
```

J. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
#Se crea un nuevo data frame que incluye unicamente los registros efectuados antes del 20 de diciembre del 2020.
antes20 = Vacunas[Vacunas['date']<'2020-12-20']
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
#Se crea un nuevo Dataframe con todos los registros donde se uso la vacuna Pfizer.
pfizer = Vacunas[Vacunas['vaccines'].str.contains('pfizer', case=False, na=False)]
```

L. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
#Se almacena en un archivo de Excel los Dataframes conDiferencias, conCantidad, antes20 y pfizer, cada uno en una hoja diferente.
archivo_excel = "/content/gdrive/MyDrive/Data Scientist/5. Panda y Numpy en Python/resultadosReto.xlsx"

with pd.ExcelWriter(archivo_excel, engine='xlsxwriter') as writer:
  conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
  conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
  antes20.to_excel(writer, sheet_name='antes20', index=False)
  pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
